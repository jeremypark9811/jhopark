 #  Daum 영화 제목 및 댓글 크롤링

```python
import requests
from bs4 import BeautifulSoup
```

### 영화 메인 정보 크롤링 함수

```python
def get_daum_movie_title(movie_id):   
    url = 'https://movie.daum.net/moviedb/main?movieId={}'.format(movie_id)
    
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text)
    movie_tag = soup.select_one('strong.tit_movie') 
    
    if movie_tag:
        return movie_tag.get_text()
    return ""
```

```python
## 영화 메인 제목 크롤링 함수 호출
movie_title = get_daum_movie_title('128372')  
```

### 영화 댓글 클롤링 함수

```python
## 영화 댓글 클롤링 함수
def get_daum_movie_reviews(movie_id, max_page):  
    url_template = 'https://movie.daum.net/moviedb/grade?movieId={}&type=netizen&page={}'
    
    page = 1
    reviews = ''
    
    while True: 
        if page > int(max_page): 
            print(">>> max page. break!")
            break
            
        url = url_template.format(movie_id, page)
        resp = requests.get(url)
        soup = BeautifulSoup(resp.text)
        
        page_reviews = soup.find_all('p', class_='desc_review') 
        page_review_length = len(page_reviews)
        
        if page_review_length == 0: 
                                    
                                    
            print(">>> page number " + str(page) + " is not exists. break!")
            break
        
        for p in soup.find_all('p', class_='desc_review'): 
            reviews += p.get_text().strip() 
            reviews += '\n' 
            
        print(">>> crawling page number : "+ str(page))
        page += 1 
    return reviews
```

```python
reviews = get_daum_movie_reviews('128372','20') 
```
